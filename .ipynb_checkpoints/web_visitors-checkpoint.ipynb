```python
#import dependencies
import os
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#import web_visitors csv file
visitors_path = os.path.join("appdata", "webvisitors_by_day_2021.csv")
visitors = pd.read_csv(visitors_path)

#remove the last sum row from the df
visitors = visitors.drop(42)

visitors
```

```python
visitors['Users'] = visitors['Users'].astype(int)
```

```python
#set fig size
plt.figure(figsize = (15,10))
#graph web visitors by day
plt.bar(visitors['Day Index'], visitors['Users'])
plt.title("User Visits per Day")
plt.xlabel("Date")
plt.ylabel("User Visits per Day")
plt.xticks(rotation=80)

plt.show()
```

```python
#read in the orders df
orders_path = os.path.join("appdata", "orders.csv")
orders_csv = pd.read_csv(orders_path)
orders_csv.head()
```

```python
orders_csv['created'].value_counts()
```

```python
orders_csv['created'] = orders_csv['created'].map(lambda x: str(x)[0:10])
```

```python
orders_csv
```

```python

orders_csv['created'] = pd.to_datetime(orders_csv['created'])
```

```python
#trim the dataframe to just include orders created in 2021
orders_21 = orders_csv['created'] > '2020-12-31'
orders_21_df = orders_csv[orders_21]
orders_21_df
```

```python
#remove the timestamp from the date
orders_21_df['created']=orders_21_df['created'].dt.date
```

```python
#plot orders per day
plt.figure(figsize=(20,10))
order_counts = pd.Series.sort_index(orders_21_df["created"].value_counts(), ascending=True)
order_counts.plot.bar()
plt.title("Orders per Day (2021)")
plt.xlabel("Date")
plt.ylabel("Orders per Day")
plt.xticks(rotation=80)

plt.show()
```
