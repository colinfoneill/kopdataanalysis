```python
#import dependencies
import os
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
#import orders csv file
orders_path = os.path.join("appdata", "orders.csv")
orders_csv = pd.read_csv(orders_path)
orders_csv.head()
```

```python
#rename the pickup address column in the orders csv to match the quotes csv
orders_csv = orders_csv.rename(columns = {'pickup.address': "pickup_address"})
```

```python
#view the dataframe
orders_csv
```

```python
#get a count of the values in the status column
orders_csv["status"].value_counts()
```

```python
#create a new column showing the difference between fee and charge amount
orders_csv['Fee/Charge Diff'] = orders_csv['fee'] - orders_csv['charge.amount']
orders_csv
```

```python
plt.figure(figsize = (15,15))
plt.scatter(orders_csv.index, orders_csv['Fee/Charge Diff'], s=10)
```

```python
#create a df of just the large differences between fee and charge amount (>200)
large_differences = orders_csv['Fee/Charge Diff'] < -200
large_differences_df = orders_csv[large_differences]
large_differences_df
```

```python
#get a count of the values in the status column for orders with large fee/charge.amount differences
large_differences_df["status"].value_counts()
```

```python
#import quotes csv file
quotes_path = os.path.join("appdata", "quotes.csv")
quotes_csv = pd.read_csv(quotes_path)
quotes_csv
```

```python
#clean the address format in the quotes dataframe to match the format in the orders dataframe
quotes_csv['pickup_address'] = quotes_csv['pickup_address'].str.split(',').str[0]
quotes_csv
```

```python
#import the custom quotes csv file
custom_quotes_path = os.path.join("appdata", "customquotes.csv")
custom_quotes_csv = pd.read_csv(custom_quotes_path)
custom_quotes_csv.count()
```

```python
#merge the quotes and orders df's
quotes_orders = pd.DataFrame.merge(orders_csv, quotes_csv, how = 'outer', on = 'pickup_address', indicator = True)
```

```python
quotes_orders
```

```python
both_quotes_orders = pd.DataFrame.merge(orders_csv, quotes_csv, how = 'inner', on = 'pickup_address', indicator = True)
both_quotes_orders
```

```python
#add column to merged dataframe showing difference between quote fee and order fee
both_quotes_orders['Fee Differences'] = both_quotes_orders['fee_x'] - both_quotes_orders['fee_y']
```

```python
#plot fee differences between orders and quotes
plt.scatter(both_quotes_orders.index, both_quotes_orders['Fee Differences'], s=10)
```

```python
#create a dataframe of just large differences (> or < abs(200))
large_fee_differences = (both_quotes_orders['Fee Differences'] > 350) | (both_quotes_orders['Fee Differences'] < -350)
large_fee_diff_df = both_quotes_orders[large_fee_differences]
large_fee_diff_df
```
