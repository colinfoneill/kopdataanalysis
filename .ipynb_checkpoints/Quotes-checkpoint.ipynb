```python
#import dependencies
import os
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
#import the regular quotes csv file
quotes_path = os.path.join('appdata', 'quotes.csv')

#read in the quotes csv
quotes = pd.read_csv(quotes_path)
```

```python
#view quotes df
quotes
```

```python
#import custom quotes csv file
custom_path = os.path.join('appdata', 'customquotes.csv')

#read in the custom quotes csv
custom = pd.read_csv(custom_path)
```

```python
#view custom_df
custom
```

```python
#create boxplot of regular and custom quotes amount
fig, ax = plt.subplots(figsize=(10,10))
ax.boxplot([quotes['amount'], custom['amount']], labels=('Regular', 'Custom'))
ax.set_title('Regular/Custom Quotes Amount')
ax.set_ylabel('Amount($)')

```
